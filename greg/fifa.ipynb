```python
import pandas as pd
import numpy as np
```

```python
# Setting low_memory=False prevents this warning in pink: 
# "DtypeWarning: Columns ___ have mixed types.  Specify dtype option on import or set low_memory=False."

fifa21 = pd.read_csv("fifa21_male2.csv", low_memory=False)
fifa21.head()
```

```python
fifa21.shape
```

```python
columns = list(fifa21.columns)  # To print all the columns
print(columns)
```

```python
# Drop unneeded columns
useless_columns = ['ID', 'Club Logo', 'Growth', 'Loan Date End', 'Release Clause', 'Contract', 'W/F', 'SM', 'A/W', \
                   'D/W', 'IR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY', 'Hits', 'LS', 'ST', 'RS', 'LW', 'LF', 'CF', \
                   'RF', 'RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', \
                   'LB', 'LCB', 'CB', 'RCB', 'RB', 'GK', 'Gender']
```

```python
fifa21 = fifa21.drop(useless_columns, axis = 1)
fifa21.head()
```

```python
fifa21.shape
```

```python
# See what Value columns looks like.  Needs some cleaning.
fifa21['Value']
```

```python
# Drop the Euro symbol
fifa21['Value'] = fifa21['Value'].str[1:]
fifa21['Value']
```

```python
# Convert Value from "625K", "5.5M", etc. to just numbers:

fifa21.Value = (fifa21.Value.replace(r'[KM]+$', '', regex=True).astype(float) * \
          fifa21.Value.str.extract(r'[\d\.]+([KM]+)', expand=False).fillna(1).replace(['K','M'], [10**3, 10**6]).astype(int))
```

```python
fifa21['Value']
```

```python
fifa21['Wage']
```

```python
# Now clean the Wage column:
# Drop the Euro sign
fifa21['Wage'] = fifa21['Wage'].str[1:]
```

```python
# Convert Value from "625K", "5.5M", etc. to just numbers:
fifa21.Wage = (fifa21.Wage.replace(r'[KM]+$', '', regex=True).astype(float) * \
          fifa21.Wage.str.extract(r'[\d\.]+([KM]+)', expand=False).fillna(1).replace(['K','M'], [10**3, 10**6]).astype(int))
```

```python
fifa21['Wage']
```

```python
fifa21['Weight']
```

```python
# Drop 'lbs' from the weight column and change it to integer
fifa21['Weight'] = fifa21['Weight'].str[:3].astype(int)
fifa21['Weight']
```

```python
# Moved this to the bottom and commented out#.  It worked in here to change height (6'0") to inches(72), but caused havoc with the csv.
def parse_ht(ht):
    # format: 7' 0.0"
    ht_ = ht.split("'")
    ft_ = float(ht_[0])
    in_ = float(ht_[1].replace("\"",""))
    return (12*ft_) + in_

fifa21['inches'] = fifa21['Height'].apply(lambda x:parse_ht(x)).astype(int)
```

```python
fifa21['inches']
```

```python
fifa21['bmi'] = (fifa21['Weight'] / (fifa21['inches']**2)) * 703
```

```python
fifa21['bmi']
```

```python
# Move the new inches column from the right side so it's next to height:
fifa21 = fifa21[['Name', 'Age', 'OVA', 'Nationality', 'Club', 'BOV', 'BP', 'Position',
       'Player Photo', 'Flag Photo', 'POT', 'Team & Contract', 'Height', 'inches',
       'Weight', 'bmi', 'foot', 'Joined', 'Value', 'Wage', 'Attacking', 'Crossing',
       'Finishing', 'Heading Accuracy', 'Short Passing', 'Volleys', 'Skill',
       'Dribbling', 'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control',
       'Movement', 'Acceleration', 'Sprint Speed', 'Agility', 'Reactions',
       'Balance', 'Power', 'Shot Power', 'Jumping', 'Stamina', 'Strength',
       'Long Shots', 'Mentality', 'Aggression', 'Interceptions', 'Positioning',
       'Vision', 'Penalties', 'Composure', 'Defending', 'Marking',
       'Standing Tackle', 'Sliding Tackle', 'Goalkeeping', 'GK Diving',
       'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes',
       'Total Stats', 'Base Stats']]
```

```python
fifa21.info()
```

```python
fifa21.columns
```

```python
# Fill NaN cells with 0 so we can perform calculations
fifa21 = fifa21.fillna(0)
```

```python
fifa21.head()
```

```python
# Calculate the attributes based on the site Kim found
# (https://www.fifauteam.com/fifa-20-attributes-guide/#)

# Pace = Acceleration + Sprint Speed
# Shooting = Finishing + Long Shots + Penalties + Positioning + Shot Power + Volleys
# Passing = Crossing +   Curve + FK Accuracy + Long Passing + Short Passing + Vision
# Dribbling = Agility + Balance + Ball Control + Composure + Dribbling + Reactions
# Defending = Heading Accuracy + Interceptions + Marking + Sliding Tackle + Standing Tackle
# Physical = Aggression + Jumping + Stamina + Strength
# Goalkeeping = GK Diving + GK Handling + GK Kicking + GK Positioning + GK Reflexes


fifa21['pace'] = fifa21['Acceleration'] + fifa21['Sprint Speed']
fifa21['shooting'] = fifa21['Finishing'] + fifa21['Long Shots'] + fifa21['Penalties'] + fifa21['Positioning'] + \
    fifa21['Shot Power'] + fifa21['Volleys']
fifa21['passing'] = fifa21['Crossing'] + fifa21['Curve'] + fifa21['FK Accuracy'] + fifa21['Long Passing'] + \
    fifa21['Short Passing'] + fifa21['Vision']
fifa21['dribbling'] = fifa21['Agility'] + fifa21['Balance'] + fifa21['Ball Control'] + fifa21['Composure'] + fifa21['Dribbling'] \
    + fifa21['Reactions']
fifa21['defending'] = fifa21['Heading Accuracy'] + fifa21['Interceptions'] + fifa21['Marking'] + fifa21['Sliding Tackle'] + \
    fifa21['Standing Tackle']
fifa21['physical'] = fifa21['Aggression'] + fifa21['Jumping'] + fifa21['Stamina'] + fifa21['Strength']
fifa21['goalkeeping2'] = fifa21['GK Diving'] + fifa21['GK Handling'] + fifa21['GK Kicking'] + fifa21['GK Positioning'] + \
    fifa21['GK Reflexes'] 
```

```python
# Drop unneeded columns and individual stat columns we don't need now that we have groups:

useless_columns = ['Position', 'Joined', 'Crossing', 'Finishing', 'Heading Accuracy', 'Short Passing', 'Volleys', \
                   'Dribbling', 'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control', 'Acceleration', 'Sprint Speed', 'Agility',\
                   'Reactions', 'Balance', 'Shot Power', 'Jumping', 'Stamina', 'Strength', 'Long Shots', 'Aggression', \
                   'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking', 'Standing Tackle', \
                   'Sliding Tackle', 'GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes',]
fifa21 = fifa21.drop(useless_columns, axis = 1)
fifa21.head()
```

```python
# Rename the columns to be sql-friendly

fifa21 = fifa21.rename(columns = {'Name':'name', 'Age':'age', 'OVA':'overall', 'Nationality':'nationality', \
        'Club':'club', 'BOV':'bov', 'BP':'bestposition',  'Player Photo':'playerphoto', 'Flag Photo':'flagphoto',\
        'POT':'potential', 'Team & Contract':'teamcontract', 'Height':'height', 'Weight':'weight', 'Value':'value',\
        'Wage':'wage', 'Attacking':'attacking', 'Skill':'skill', 'Movement':'movement', 'Power':'power', \
        'Mentality':'mentality', 'Defending':'defending', 'Goalkeeping':'goalkeeping', 'Total Stats':'totalstats',\
                                 'Base Stats':'basestats'})
```
