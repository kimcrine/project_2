```python
import pandas as pd
import numpy as np
```

```python
# Setting low_memory=False prevents this warning in pink: 
# "DtypeWarning: Columns ___ have mixed types.  Specify dtype option on import or set low_memory=False."
# Read in the csv:

fifa21 = pd.read_csv("fifa21_male2.csv", low_memory=False)
fifa21.head()
```

```python
fifa21.shape
```

```python
# Drop the players that are rated less than 75:

fifa21 = fifa21[fifa21['OVA'] >=75]
```

```python
# See how many rows left:

fifa21.shape
```

```python
columns = list(fifa21.columns)  # To print all the columns
print(columns)
```

```python
# Gather unneeded columns to drop: 

useless_columns = ['ID', 'Club Logo', 'Growth', 'Loan Date End', 'Release Clause', 'Contract', 'W/F', 'SM', 'A/W', \
                   'D/W', 'IR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY', 'Hits', 'LS', 'ST', 'RS', 'LW', 'LF', 'CF', \
                   'RF', 'RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', \
                   'LB', 'LCB', 'CB', 'RCB', 'RB', 'GK', 'Gender']
```

```python
# Drop the unneeded columns:

fifa21 = fifa21.drop(useless_columns, axis = 1)
fifa21.head()
```

```python
# See what Value columns looks like.  Needs some cleaning.
fifa21['Value']
```

```python
# Drop the Euro symbol
fifa21['Value'] = fifa21['Value'].str[1:]
fifa21['Value']
```

```python
# Convert Value from "625K", "5.5M", etc. to just numbers:

fifa21.Value = (fifa21.Value.replace(r'[KM]+$', '', regex=True).astype(float) * \
          fifa21.Value.str.extract(r'[\d\.]+([KM]+)', expand=False).fillna(1).replace(['K','M'], [10**3, 10**6]).astype(int))
```

```python
fifa21['Value']
```

```python
# Look at the Wage column:
fifa21['Wage']
```

```python
# Now clean the Wage column:
# Drop the Euro sign
fifa21['Wage'] = fifa21['Wage'].str[1:]
```

```python
# Convert Value from "625K", "5.5M", etc. to just numbers:
fifa21.Wage = (fifa21.Wage.replace(r'[KM]+$', '', regex=True).astype(float) * \
          fifa21.Wage.str.extract(r'[\d\.]+([KM]+)', expand=False).fillna(1).replace(['K','M'], [10**3, 10**6]).astype(int))
```

```python
# Look at Wage again:

fifa21['Wage']
```

```python
# Look at the Weight column:
fifa21['Weight']
```

```python
# Drop 'lbs' from the weight column and change it from a string to integer:

fifa21['Weight'] = fifa21['Weight'].str[:3].astype(int)
fifa21['Weight']
```

```python
# Change Height to inches:

def parse_ht(ht):
    # format: 7' 0.0"
    ht_ = ht.split("'")
    ft_ = float(ht_[0])
    in_ = float(ht_[1].replace("\"",""))
    return (12*ft_) + in_

fifa21['inches'] = fifa21['Height'].apply(lambda x:parse_ht(x)).astype(int)
```

```python
fifa21['inches']
```

```python
# Calculate a bmi in case we want to analyze players that way:

fifa21['bmi'] = ((fifa21['Weight'] / (fifa21['inches']**2)) * 703).round(1)
```

```python
fifa21['bmi']
```

```python
# Move the new inches column from the right side so it's next to height:

fifa21 = fifa21[['Name', 'Age', 'OVA', 'Nationality', 'Club', 'BOV', 'BP', 'Position',
       'Player Photo', 'Flag Photo', 'POT', 'Team & Contract', 'Height', 'inches',
       'Weight', 'bmi', 'foot', 'Joined', 'Value', 'Wage', 'Attacking', 'Crossing',
       'Finishing', 'Heading Accuracy', 'Short Passing', 'Volleys', 'Skill',
       'Dribbling', 'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control',
       'Movement', 'Acceleration', 'Sprint Speed', 'Agility', 'Reactions',
       'Balance', 'Power', 'Shot Power', 'Jumping', 'Stamina', 'Strength',
       'Long Shots', 'Mentality', 'Aggression', 'Interceptions', 'Positioning',
       'Vision', 'Penalties', 'Composure', 'Defending', 'Marking',
       'Standing Tackle', 'Sliding Tackle', 'Goalkeeping', 'GK Diving',
       'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes',
       'Total Stats', 'Base Stats']]
```

```python
fifa21.info()
```

```python
# Fill NaN cells with 0 so we can perform calculations
fifa21 = fifa21.fillna(0)
```

```python
fifa21.head()
```

```python
# Max scores for aggregate categories on original CSV:
# Attacking: 437, Skill: 470, Movement: 464, Power: 444, Mentality: 421, Defense: 272, Goalkeeping: 440

# Calculate new aggregates based on scores out of 100 so that they make more sense:

fifa21['attacking'] = (fifa21['Attacking'] / 4.37).round(1)
fifa21['skill'] = (fifa21['Skill'] / 4.70).round(1)
fifa21['movement'] = (fifa21['Movement'] / 4.64).round(1)
fifa21['power'] = (fifa21['Power'] / 4.44).round(1)
fifa21['mentality'] = (fifa21['Mentality'] / 4.21).round(1)
fifa21['defending'] = (fifa21['Defending'] / 2.72).round(1)
fifa21['goalkeeping'] = (fifa21['Goalkeeping'] / 4.40).round(1)
fifa21['attacking']
```

```python
# See if the max score for each skill group is 100 now:

print(fifa21['attacking'].max())
print(fifa21['skill'].max())
print(fifa21['movement'].max())
print(fifa21['power'].max())
print(fifa21['mentality'].max())
print(fifa21['defending'].max())
print(fifa21['goalkeeping'].max())
```

```python
# Change to whole numbers, like Overall scores are:

fifa21['attacking'] = fifa21['attacking'].astype(int)
fifa21['skill'] = fifa21['skill'].astype(int)
fifa21['movement'] = fifa21['movement'].astype(int)
fifa21['power'] = fifa21['power'].astype(int)
fifa21['mentality'] = fifa21['mentality'].astype(int)
fifa21['defending'] = fifa21['defending'].astype(int)
fifa21['goalkeeping'] = fifa21['goalkeeping'].astype(int)
fifa21['attacking']
```
