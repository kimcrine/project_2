```python
import pandas as pd
import json
```

```python
fifa = pd.read_csv("fifalarge.csv", low_memory=False)
fifa.head()
```

```python
fifa.shape
```

```python
#count total purchases for future calculations
total = fifa["nationality"].count()

#find unique purchasers
unique_country = fifa["nationality"].unique()

#calculate total numbers of players
num_countries=len(unique_country)
num_players=total

print("Total Countries = ",num_countries)
print("Total Players = ",num_players)
```

```python

# Use GroupBy to separate the data into fields according to "Gender" values
grouped_df = fifa.groupby(['nationality'])


#count purchasers
player_count = grouped_df["nationality"].count()

#calculate total price
avg_overall = grouped_df["overall"].mean()

#calculate average price
#g_avg = g_price / grouped_df["SN"].count()

#calcualte total purchase per person
#g_pur = g_price / unique_users["Gender"].value_counts()

#set dictionary for proper formatting
#format_dict = {'Player Count':'{:.0f}', 'Average Overall': '${:,.0f}'}

#create dataframe for output
fifa_countries = pd.DataFrame({"Player Count":player_count,
                                "Average Overall": avg_overall})

#output df in proper format
#fifa_countries.style.format(format_dict)
high_count=fifa_countries["Player Count"].max()

fifa_countries["Weighted Count"]=(fifa_countries["Player Count"]/high_count)*100



fifa_countries


```

```python
#{ country: "Brazil", color: "#009900", score: "95" },
#  { country: "Argentina", color: "#7ade45", score: "85" },
#  { country: "Colombia", color: "#a3e84f", score: "75" },
#  { country: "Uruguay", color: "#d9f000", score: "65" },
#  { country: "Paraguay", color: "#ffff00", score: "55" },
#  { country: "Chile", color: "#fcf200", score: "45" },
#  { country: "Peru", color: "#ffcc52", score: "35" },
#  { country: "Venezuela", color: "#ff993d", score: "25" },
#  { country: "Ecuador", color: "#ff8033", score: "15" },
#  { country: "Bolivia", color: "#cc0000", score: "5" }];





#color_name
colors = ["#009900","#7ade45","#a3e84f","#d9f000","#ffff00","#fcf200","#ffcc52","#ff993d","#ff8033","#cc0000"]


# set series empty to append
color_list =[]

#  count the unique mice per drug regimen
for i in fifa_countries["Weighted Count"]:
    if i >= 90:
        color_list.append(colors[0])
    elif i >= 80:
        color_list.append(colors[1])
    elif i >= 70:
        color_list.append(colors[2])
    elif i >= 60:
        color_list.append(colors[3])
    elif i >= 50:
        color_list.append(colors[4])
    elif i >= 40:
        color_list.append(colors[5])
    elif i >= 30:
        color_list.append(colors[6])
    elif i >= 20:
        color_list.append(colors[7])
    elif i >= 10:
        color_list.append(colors[8])
    else:
        color_list.append(colors[9])   
    
    
# create a datafram of unique mice per drug without duplicates for graphing
#drug_mice = pd.DataFrame({"Drug Regimen":unique_mice.index,
 #                               "Number of Mice": mice_pd})

# create a datafram of unique mice per drug without duplicates for graphing
#drug_mice = pd.DataFrame({"Drug Regimen":unique_mice.index,
 #                               "Number of Mice": mice["Mouse ID"].count()})

fifa_countries["Color"]=color_list



# sort Top spenders from highest to lowest purchase value
fifa_countries = fifa_countries.sort_values("Player Count", ascending=False)

fifa_countries



fifa_countries



#fifa_countries["Weighted Count"]=(fifa_countries["Player Count"]/high_count)*100
```
