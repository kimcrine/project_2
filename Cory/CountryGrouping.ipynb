```python

country_list=[]

for item in fifa["nationality"]:
    if item == "Antigua &amp; Barbuda":
        country_list.append("Antigua and Barbuda")
    elif item == "Bosnia Herzegovina":
        country_list.append("Bosnia and Herzegovina")
    elif item == "Brunei Darussalam":
        country_list.append("Brunei")
    elif item == "China PR":
        country_list.append("China")    
    elif item == "Chinese Taipei":
        country_list.append("China")    
    elif item == "Hong Kong":
        country_list.append("China")    
    elif item == "Curacao":
        country_list.append("Curaçao")    
    elif item == "DR Congo":
        country_list.append("Democratic Republic of the Congo")    
    elif item == "Republic of Ireland":
        country_list.append("Ireland")    
    elif item == "Macau":
        country_list.append("Macao S.A.R")    
    elif item == "North Macedonia":
        country_list.append("Macedonia")
    elif item == "Korea DPR":
        country_list.append("North Korea")    
    elif item == "Congo":
        country_list.append("Republic of Congo")    
    elif item == "Serbia":
        country_list.append("Republic of Serbia")    
    elif item == "São Tomé &amp; Príncipe":
        country_list.append("Sao Tome and Principe")
    elif item == "Korea Republic":
        country_list.append("South Korea")
    elif item == "Trinidad &amp; Tobago":
        country_list.append("Trinidad and Tobago")
    elif item == "England":
        country_list.append("United Kingdom")
    elif item == "Scotland":
        country_list.append("United Kingdom")
    elif item == "Wales":
        country_list.append("United Kingdom")
    elif item == "Northern Ireland":
        country_list.append("United Kingdom")
    elif item == "Tanzania":
        country_list.append("United Republic of Tanzania")
    elif item == "United States":
        country_list.append("United States of America")    
    else: 
        country_list.append(item)
        
country_list

```

```python
fifa["nationality"]=country_list

fifa
```

```python
top10000 = fifa.head(10000)

top10000
```

```python
#count total purchases for future calculations
total = top10000["nationality"].count()

#find unique purchasers
unique_country = top1000["nationality"].unique()

#calculate total numbers of players
num_countries=len(unique_country)
num_players=total

print("Total Countries = ",num_countries)
print("Total Players = ",num_players)
```

```python

# Use GroupBy to separate the data into fields according to "Gender" values
grouped_df = top10000.groupby(['nationality'])


#count purchasers
player_count = grouped_df["nationality"].count()

#calculate total price
avg_overall = grouped_df["overall"].mean()

#calculate average price
#g_avg = g_price / grouped_df["SN"].count()

#calcualte total purchase per person
#g_pur = g_price / unique_users["Gender"].value_counts()

#set dictionary for proper formatting
#format_dict = {'Player Count':'{:.0f}', 'Average Overall': '${:,.0f}'}

#create dataframe for output
fifa_countries = pd.DataFrame({"Player Count":player_count,
                                "Average Overall": avg_overall})

#output df in proper format
#fifa_countries.style.format(format_dict)
high_count=fifa_countries["Player Count"].max()

fifa_countries["Weighted Count"]=(fifa_countries["Player Count"]/high_count)*100



fifa_countries


```
